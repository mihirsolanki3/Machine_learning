```python
# Data for the pie chart
stud= np.array([50,60,30,40,20])# Number of students in each track
trac_name=["Data","Mobile","CyberSecurity","Biotechnolgy","IoT"]# Names of the tracks
ex = [0.2, 0, 0, 0, 0]  #slice (Data)
col = ["Red", "Pink", 'm', 'g', 'b']  # Colors for each slice
plt.pie(stud,labels=trac_name,startangle=90,explode=ex,shadow=True,colors=col)# Create the pie chart
plt.legend(title="Tracks")# Add a legend with a title
plt.show()# display the pie chart
```
